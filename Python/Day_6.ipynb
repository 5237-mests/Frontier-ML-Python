### **Step 1: Create a table**

First, we'll set up a table named books that will store the title, author, and publication year of the books.

```python
import sqlite3

# Connect to a database (or create one if it doesn't exist)
conn = sqlite3.connect('library.db')
c = conn.cursor()

# Create a table named 'books'
c.execute('''
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY,
    title TEXT,
    author TEXT,
    year INTEGER
)
''')
#save all the changes made in database
conn.commit()
#close connection
conn.close()
```

### **Step 2: Insert a row**

Now, we'll insert a sample book into the table.

* **Insert a few rows (records)**

```python
conn = sqlite3.connect('library.db') #reconnect to database

c = conn.cursor() #execute SQL commands

# Insert a few books
#c.execute(SQL command, tuple of values)
c.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)", ("To Kill a Mockingbird", "Harper Lee", 1960))
c.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)", ("The Power of Now", "Ecklart Tolle", 2004))
c.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)", ("The Miracle Morning", "Hal Elrod", 2012))

conn.commit() #save
conn.close() #close connection
```

# **Introduction to Relationships between Tables**

* In the real world, data often has relationships.

 * For example, one student might enroll in multiple courses, and each course might have multiple students.

* In the digital database world, relationships are meaningful associations between tables that contain related information.

* Without some connection between tables in a database, you may as well be working with disparate spreadsheet files rather than a database system.



**Recall:**
* Databases are collections of tables, and those tables have fields (columns) and records (rows)

* Relational databases use  keys to establish relationships (connections) between tables. Also to ensure data integrity and avoid redundancy.

* By telling your database that the key values in one table correspond to key values in another, you create a relationship between those tables.

* These relationships make it possible to run powerful queries across different tables in your database.

## **Understanding Keys in Relational Database Tables**



[Image Source](https://www.google.com/url?sa=i&url=https%3A%2F%2Fmacxima.medium.com%2Fprimary-key-vs-foreign-key-in-sql-d89260b74e1c&psig=AOvVaw198H0vUZnEOCKXXJidVF13&ust=1693107827611000&source=images&cd=vfe&opi=89978449&ved=0CBIQjhxqFwoTCKjAq6i0-YADFQAAAAAdAAAAABAJ)

Three fundamental keys of databases: primary, foreign, and unique keys.

### **1. Primary Keys**

* A unique identifier for each record in a table. No duplicates or nulls allowed.
* Every table contains a primary (entity) key, which identifies the rows within that table.
* For instance, each student in a school might have a specific "Student ID" that no other student has. This "Student ID" could serve as our primary key.



### **2. Foreign Keys**

* A field in one table that uniquely identifies a record in another table, thereby establishing a relationship between them.

* When one table’s entity key gets linked to a second table, it’s known as a foreign key in that second table.


>When a primary key appears in another table, that field is called a foreign key in that table.
>The relationship between primary keys and foreign keys creates the relationship between tables.

### **3. Unique Keys**

* It ensures that the value is unique across the table (where it has been set).
* Unlike primary keys, unique keys can have null (or empty) values, and a table can have more than one unique key.


**Note:**

* Identifying the connections you will  need between tables is part of the data modeling and schema design process (see later)

## **Relationships in Tables**


* To represent relationships in databases, we use the following terms:


## **1. One-to-One Relationship**



**Real-world example:**

>Every human has a unique ID depending on the country. In this analogy, one person relates to one and only one ID number, and vice versa.

**Database context:**
>This relationship implies that a record in one table (Table A) relates to only one record in another table (Table B), and a record in Table B relates to only one record in Table A.

**Example:**
>Consider two tables: Users and Licenses. Each user has only one driver's license, and each driver's license belongs to only one user. Here, a record in the Users table relates to a single record in the Licenses table, establishing a one-to-one relationship.

* One-to-one relationships can be also useful for security purposes. For instance: You may want to store sensitive customer information in a separate table, which you link to your main Customers table with a foreign key.



## **2. One-to-Many Relationship:**

**Real-world example:**
>A mother can have multiple children, but each child has one biological mother. Here, one mother relates to many children, but each child relates to one mother.

**Database context:**
>A record in Table A can relate to many records in Table B, but a record in Table B relates to only one record in Table A.


**Example:**
>Consider two tables: Authors and Books. An author can write many books, but each book has only one primary author. In this case, a record in the Authors table can relate to multiple records in the Books table, establishing a one-to-many relationship.


* One-to-many relationships are the most common type of relationships between tables in a database.



##3. **Many-to-Many Relationship:**

**Real-world example:**
>In a school, a student can enroll in multiple courses, and a course can have multiple students. One student relates to many courses, and one course relates to many students.

**Database context:**
>Records in Table A can relate to multiple records in Table B, and vice versa.

**Example:**
>Consider two tables: Students and Courses. Since a student can enroll in multiple courses and a course can have multiple students, there exists a many-to-many relationship between them. However, relational databases can't directly represent many-to-many relationships. So, we introduce a third table, often called a junction or bridge table, to resolve this.

For the students and courses example, the third table could be StudentCourses with two primary fields: StudentID and CourseID. Each combination of StudentID and CourseID represents a student enrolled in a specific course. This table helps to break down the many-to-many relationship into two one-to-many relationships: one between Students and StudentCourses, and the other between Courses and StudentCourses.

# **Designing a Relational Database [Case Study - Retailer]**

We will walk through the design steps of a database for a retailer that wants to manage their products, customers, orders, and employees.

**Objective:**

> Design a database for a new retailer in town.

**Scenario**
> This retailer really cares about customer relationships and wants to reward customers who meet a spending goal and gift these top customers on the 1 year anniversary of their first purchase. This retailer needs a way to organize products by price and category to make smart recommendations to their customers based on their age. This retailer also wants to track the best-performing employees to reward those with the highest sales with a raise at the end of the year.

* Designing a relational database involves a structured approach to ensure that the data is well-organized, maintains its integrity, and can be accessed efficiently.

* The design process involves data modelling and schema design

* The data model is the initial, high-level representation of the data requirements and relationships.
 * It provides an abstract view of the data elements, their relationships, and other constraints without delving into specific database system details.

* The schema is the structure that we define for our data. The schema defines the tables, relationships between tables, fields, and indexes.




* The data modeling and schema design process allows you to figure out how your data fits together, and how exactly you should configure your tables and their fields.


* One tool that will help us design our schema is an ERD (entity-relationship diagram).
 * We'll use [Lucidchart to build out our ERD](https://www.lucidchart.com/pages/), and you can sign up for free.

* This diagram will allow us to visualize our entities and their relationships.

Here are the major to-dos when designing our schema:

* Understand business needs
* Identify entities
* Identify properties/fields on those entities
* Define relationships between tables

### **Step 1: Understand Business Needs**

* In designing a relational database schema, first step is to understand the needs of the business.
* Help us determine what type of information we should be storing.
  * For example, if we are working with a retailer that wants to offer an anniversary gift for clients on their first anniversary, we would have to store the date a customer joins.

A recap of the requirements for our customers:

* Store customer spending to-date
* Store customer anniversary date of first purchase
* Store customer’s age
* Store employee sales total in dollar amount
* Store products and include a category and price property

### **Step 2: Defining Entities (Tables)**

* The next step is defining the entities we want to store data about.

* In our retailer case, entities should be:

  * customers
  * products
  * orders
  * employees

* Using the ERD tool, we name each entity.

> Note: This could extend to add more entities like stores if there are multiple storefront locations, manufacturers, etc., depending on the needs of the business.

image.png

### **Step 3: Defining Properties**

* Once we identify entities, we should define what fields we want to store about these entities.
* Note: each table, or entity, should have one unique, identifying property **(primary key)**.
 * For example, we may have customers with the same name or same birthdate.

* Two common ways to come up with a primary key:
 * Programmatically generate a unique value
 * Assign an integer that automatically increases with each new entry





image.png

* Everything in tables properties comes directly from the specs that the business gave us.
 * For example, the business wants to know which customer made the purchase, which employee made the sale, and which products were in the order.
 Notice in Orders table, a customerID, employeeID, and productID is referenced to meet those needs.

### **Step 4: Define Relationships**

* After identifying entities and their properties, think about how these tables relate to each other.
* A parent table will have a unique primary key column, and a child table will have its own primary key and then a parent_id column that references the parent table.


image.png


>For example, the customers table has a customerID, which is the primary key. In the Orders table, we set an orderID as the primary key and reference the customerID to denote which customer made the order. Similarly, we also have a column referencing the Employees table, employeeID, to denote which employee made the sale.
>When a primary key appears in another table, that field is called a foreign key in that table. The relationship between primary keys and foreign keys creates the relationship between tables.
