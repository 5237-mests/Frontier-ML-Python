## **What is a Search Engine?**

A search engine is a web-based software system that can be used to look for various information on the internet.

* It is an efficient and powerful internet tool to comfortably collect required information.

* It is an automated software application that can be used to perform several functions such as Crawling, Indexing, Storing Information, and Giving results.

* A search engine consists of two main components, a database of indexed links and files, and a series of algorithms that search for and return query results.

## **Part 1: Web Crawlers**

Web Crawling (spidering)  is the process of systematically browsing the web to collect information about websites and their pages.

* The bot or spiders or crawlers are sent across different websites simultaneously to collect a large amount of information.


* This information can then be used by search engines to index and rank these pages.

### **Objective of Web Crawling**

The main objectives of web crawling are:

* To fetch web page data so it can be indexed by search engines.
* To update existing content in the search engine database with newer versions of web pages.
* To discover new pages or websites.


image.png

### **Basic Components of a Web Crawler**
* Seed URLs: A list of initial URLs to start the crawling process.
* URL Frontier: A queue of URLs waiting to be fetched.
* HTTP Fetcher: A component to make HTTP requests and retrieve page content.
* Content Extractor: Parses the page to extract useful content and more URLs.
* Data Storage: A database or file system to store crawled data.
* Crawling Policies: Rules that determine which pages to crawl, how often, and the depth.

### **Steps in Web Crawling**
* Start with Seed URLs: These are initial URLs that the crawler starts with (e.g. Wikipedia.com, WSJ.com, NYT.com).
* Fetch the Web Page: Using HTTP requests, the content of the URL is fetched.
* Parse the Page: Extract useful information and other linked URLs.
* Store the Content: Save the content in a database for indexing and future reference.
* Add New URLs to the Queue: URLs extracted from the page are added to the URL Frontier.
* Check for **Robots.txt**: Respect website's robots.txt rules which specify which pages or resources shouldn't be crawled (e.g google.com/robots.txt).
* Repeat: Continue the process with the next URL in the frontier until the desired number of pages are crawled or the frontier is empty.


### **Challenges in Web Crawling**
* Large Volume of Data: The web is vast, and even a small fraction can be millions of pages.
* Dynamic Content: Web pages can change, and new content is added frequently.
* Deduplication: Avoiding crawling the same page multiple times.
* Crawl Policies: Respecting robots.txt and not overloading servers with frequent requests.
* Handling Different Data Types: Not all content is HTML; there are PDFs, images, videos, etc.


### **Tools and Libraries for Web Crawling**

* BeautifulSoup: A Python library for parsing HTML and XML documents.
* Scrapy: An open-source framework for web crawling and scraping.
* Requests: A Python library for making HTTP requests.

##**Objective of Indexing**
The primary objective of indexing is:

* To create a data structure that allows for quick retrieval of relevant documents based on search queries.








## **List of Common Indices**


### **1. Forward Index**

* A forward index is a data structure used in information retrieval systems, particularly in search engines.

* It maps from documents to the terms (or words) they contain. I

* That's for each document in a collection, the forward index lists all the terms that appear in that document along with their associated metadata (like frequency, position, etc.).




image.png


### **2. Inverted Index**

* An inverted index is a data structure used to map terms (words) to their locations in documents.
* In inverted index, the index is organized by terms (words), and each term points to a list of documents or web pages that contain that term.

* Similar to index at the back of books.

* Used in information retrieval systems to efficiently retrieve documents or web pages containing a specific term or set of terms. In an inverted index, the index is organized by terms (words), and each term points to a list of documents or web pages that contain that term.



image.png




## **Steps in Building an Inverted Index**
* Tokenization: Split the content of each document into individual words or terms.
* Normalization: Convert all terms to a standard form, e.g., lowercasing, stemming, or lemmatization.
* Term Frequency Calculation: Count how often each term appears in a document.
* Document List Creation: For each term, create a list of documents where it appears.
* Storing the Index: Store the inverted index in a database or a file system for fast retrieval.


## **Tools and Libraries for Indexing**

* Lucene: A high-performance, full-featured text search engine library.
* Elasticsearch: A search and analytics engine built on top of Lucene.
* Whoosh: A fast, featureful full-text indexing and searching library implemented in pure Python.

## **Part 3. PageRank Ranking Algorithm**










image.png


Algorithm by Larry Page and Sergey Brin, the founders of Google, designed to rank web pages in search engine results by considering the link structures of the web.
* Basically, it determines the authoritativeness of a page on the internet.

* For example, when you search *Earth Day*, we need to know how trustworthy is the page. Otherwise, search engine could send them to a random blog page of *earth day*, instead of *Wikipedia* or *earthday.com*

### **Objective of PageRank**
* To measure the importance of web pages based on their inbound links.
* To provide a foundation for search engines to rank web pages in search results.


### **How Does PageRank Work?**

At its core, PageRank views the web as a directed graph, where web pages are nodes and hyperlinks are directed edges between nodes.
* A link from page A (source URL) to page B (destination URL) is treated as a "vote" from A to B.


###**Basic Principles:**

* Every page has an initial, equal rank.
* Each page distributes its rank over its outbound links.
* A page's rank is updated based on the ranks of pages linking to it.
